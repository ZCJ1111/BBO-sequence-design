```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_file(path):
    df = pd.read_csv(path)
    
    value = df['scoremax'].values
    avg = np.mean(value)
    std = np.std(value)
    return value, avg, std
```

```python
get_file('result/batchbo-cnn-1ADQ_seed1.csv')
```

```python
rounds=np.arange(0,40)
value_pex, avg_pex, std_pex = get_file('result/pex-cnn-1ADQ_seed1.csv')
plt.plot(rounds,value_pex,label='PEX',color='red')
plt.fill_between(rounds, avg_pex+std_pex,avg_pex-std_pex,facecolor='red',alpha=0.1)

value_bo, avg_bo, std_bo = get_file('result/batchbo-cnn-1ADQ_seed1.csv')
plt.plot(rounds,value_bo,label='BatchBO',color='green')
plt.fill_between(rounds, avg_bo+std_bo,avg_bo-std_bo,facecolor='green',alpha=0.1)

value_ran, avg_ran, std_ran = get_file('result/random-cnn-1ADQ_seed1.csv')
plt.plot(rounds,value_ran,label='Random',color='blue')
plt.fill_between(rounds, avg_ran+std_ran,avg_ran-std_ran,facecolor='blue',alpha=0.1)

plt.ylabel('Max fitscore')
plt.legend()
plt.xlabel('Rounds')
plt.title('1ADQ')
```
